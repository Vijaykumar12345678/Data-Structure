1.Write a  python program to implement the stack.

```python
class Stack:
    def __init__(self):
        # Initialize an empty list to store stack elements
        self.stack = []

    def push(self, item):
        """
        Description:
            This function is used to append an element into the stack.
        
        Parameters:
            Item: str to append and element.
            
        Returns:
            None"""
        self.stack.append(item)

    def pop(self):
        
        """
        Description:
            This function  is used to remove the element from the stack and if its empty it will display
            the stack is empty
        
        Parameters:
            None
        
        Returns:
            stack or str
        
        """
        
        if not self.is_empty():
            return self.stack.pop()
        else:
            return "Stack is empty"

    def peek(self):
        
        """
        Description:
            This function is used to display the top element in the stack
            
        
        Parameters:
            None
        
        Returns:
           int or str
        """
        
        if not self.is_empty():
            return self.stack[-1]
        else:
            return "Stack is empty"

    def is_empty(self):
        """
        Description:
            This function is used to check the stack is empty or not
        
        Parameters:
            None
        
        Returns:
            Int: length of the stack"""
        return len(self.stack) == 0

    def size(self):
        """
        Description:
            This function is used to find the size of the stack.
        
        Parameters:
            None
        
        Returns:
            Int: length of the stack"""
        return len(self.stack)
    def display(self):
        """
        Description:
            This function is used to display the elements in the  stack
        
        Parameters:
            None
        
        Returns:
            stack"""
        stack=self.stack
        return stack

def main():
    s = Stack()
    
    while True:
        print("\nStack Operations:")
        print("1. Push an element")
        print("2. Pop an element")
        print("3. Peek the top element")
        print("4. Display the elements n the stack")
        print("5. Check if stack is empty")
        print("6. Get stack size")
        print("7. Exit")
        
        choice = int(input("Enter your choice (1-7): "))
        
        if choice == 1:
            item = input("Enter the element to push: ")
            s.push(item)
            print(f"Pushed {item} onto the stack.")
        elif choice == 2:
            popped_item = s.pop()
            print(f"Popped item: {popped_item}")
        elif choice == 3:
            top_item = s.peek()
            print(f"Top element is: {top_item}")
        elif choice==4:
            print(s.display())
        elif choice == 5:
            if s.is_empty():
                print("Stack is empty.")
            else:
                print("Stack is not empty.")
        elif choice == 6:
            print(f"Stack size is: {s.size()}")
        elif choice == 7:
            print("Exiting...")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

2.Write a python program on Check for Balanced Brackets in an expression (well-formedness)

```python
def brackets_balanced(expression):
    """
    Description:
        Checks whether the brackets in the given expression are balanced.

    Parameters:
        expression : str A string containing the expression with brackets.

    Returns:
        bool: True if the expression has balanced brackets, False otherwise.


    """
    stack = []
    for ch in expression:
        if ch in ('(', '{', '['):
            stack.append(ch)
        else:
            if stack and ((stack[-1] == '(' and ch == ')') or
                          (stack[-1] == '{' and ch == '}') or
                          (stack[-1] == '[' and ch == ']')):
                stack.pop()
            else:
                return False
    return not stack


def main():
   
    expression = input("Enter the expression: ")
    if brackets_balanced(expression):
        print("Balanced")
    else:
        print("Not Balanced")


if __name__ == "__main__":
    main()
```

3. Write a python program to reverse a string using stack.

```python


def isEmpty(stack):
    """
    Description:
        This function is used to check the stack is empty or not.
    parameters:
    
        stack:
    
    Returns:
        Bool
    """
    if len(stack) == 0:
        return True

def push(stack, item):
    """
    Description:
        This function is used to append an to the stack.

    parameters:
        stack: The stack where the item will be added.
        item:The item which has  to be added to the stack.

    
    Returns:
        None
    """    
    stack.append(item)


def pop(stack):

    """
    Description:
        This function is used to pop and element from the stack.

    parameters:
        stack: The stack where the item will be added.
      

    
    Returns:
        None
    """ 
    if isEmpty(stack):
        return
    return stack.pop()



def reverse(string):

    """
    Description:
        This function is used to reverse a string.

    parameters:
        string: The given user string

    
    Returns:
        string
    """ 
    n = len(string)
    stack = []
    for i in range(0, n, 1):
        push(stack, string[i])
    string = ""
    for i in range(0, n, 1):
        string += pop(stack)
    
    return string

def main():
    string = input("Enter the string :")
    string = reverse(string)
    print("Reversed string is " + string)

if __name__=="__main__":
    main()
```

7.Write a python program on to implement the binary tree.

```python
class Node:
   
    def __init__(self, data):
        self.data = data  
        self.left = None  
        self.right = None  

class BinaryTree:
    
    def insert(self, root, data):
        """
        Description
            Inserts a new node with the given data into the binary tree.

        Parameters:
            root (Node): The root node of the binary tree.
            data (Any): The value to be inserted into the tree.
        
        Returns:
            Node: The root node after insertion.
        """
        if root is None:
            return Node(data)
        else:
            if data < root.data:
                root.left = self.insert(root.left, data)
            else:
                root.right = self.insert(root.right, data)
        return root

    def inorder_traversal(self, root):
        """
        Description:
            Performs in-order traversal (left, root, right) of the binary tree.

        Parameters:
            root (Node): The root node of the binary tree.
        
        Returns:
            None
        """
        if root:
            self.inorder_traversal(root.left)
            print(root.data, end=" ")
            self.inorder_traversal(root.right)

    def preorder_traversal(self, root):
        """
        Description:
            Performs pre-order traversal (root, left, right) of the binary tree.

        Parameters:
            root (Node): The root node of the binary tree.
        
        Returns:
            None
        """
        if root:
            print(root.data, end=" ")
            self.preorder_traversal(root.left)
            self.preorder_traversal(root.right)

    def postorder_traversal(self, root):
        """
        Description:
            Performs post-order traversal (left, right, root) of the binary tree.

        parameters:
            root (Node): The root node of the binary tree.
        
        Returns:
            None

        """
        if root:
            self.postorder_traversal(root.left)
            self.postorder_traversal(root.right)
            print(root.data, end=" ")


def main():
    tree = BinaryTree()
    root = None
    
   
    values = [50, 30, 70, 20, 40, 60, 80]
    for value in values:
        root = tree.insert(root, value)

    print("In-order traversal:")
    tree.inorder_traversal(root)  

    print("\nPre-order traversal:")
    tree.preorder_traversal(root)  

    print("\nPost-order traversal:")
    tree.postorder_traversal(root)  

if __name__ == "__main__":
    main()
```

8. Write a Python program on Maximum Depth or Height of Binary Tree

```python
class Node:
   
    def __init__(self, data):
        self.data = data  
        self.left = None  
        self.right = None  


class BinaryTree:
    def insert(self, root, data):
        """
        Description:
            Inserts a new node with the given data into the binary tree.

        Parameters:
            root (Node): The root node of the binary tree.
            data (Any): The value to be inserted into the tree.
        
        Returns:
            Node: The root node after insertion.
        """
        if root is None:
            return Node(data)
        else:
            if data < root.data:
                root.left = self.insert(root.left, data)
            else:
                root.right = self.insert(root.right, data)
        return root

    
    def max_depth(self, node):
        """
        Description:
            Calculates the maximum depth of the binary tree.

        Parameters:
            node (Node): The node from where the depth calculation begins.

        Returns:
            int: The maximum depth of the tree.
        """
        if node is None:
            return 0
        else:
           
            left_depth = self.max_depth(node.left)
            right_depth = self.max_depth(node.right)

          
            return max(left_depth, right_depth) + 1


def main():
   
    tree = BinaryTree()
    root = None

    values = [50, 30, 70, 20, 40, 60, 80,10]
    for value in values:
        root = tree.insert(root, value)
    print("\n Maximum depth of the tree:", tree.max_depth(root))  


if __name__ == "__main__":
    main()
```
